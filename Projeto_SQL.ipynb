## **1.**Sobre a base de dados

Os dados contém informações sobre os clientes de um banco, organizados nas seguintes colunas:

*   **idade:** idade do cliente;
*   **sexo:**  gênero do cliente (feminino ou masculino);
*   **dependentes:** quantidade de filhos;
*   **escolaridade:** nível de escolaridade;
*   **salario_anual:** faixa salarial anual do cliente;
*   **tipo_cartao:** modalidade de cartão do cliente;
*   **qtd_produtos:** quantidade de produtos comprados nos últimos 12 meses;
*   **iteracoes_12m:** quantidade de transações nos últimos 12 meses;
*   **meses_inativo_12m:** quantidade de meses que o cliente ficou inativo;
*   **limite_credito:** limite de crédito fornecido ao cliente;
*   **valor_transacoes_12m:** valor total das transações realizadas nos últimos 12 meses;
*   **qtd_transacoes_12m:** quantidade de transações realizadas nos últimos 12 meses.

##**4.** Conclusão

Podemos extrair algumas conclusões interessantes sobre esta base de dados, como por exemplo:
*    A maior parte dos clientes deste banco possui uma menor renda anual (até 40K); 
*   Não existe uma faixa etária predominante entre os clientes;
*   A escolaridade parece não influenciar no limite de crédito obtido;
*   Dentre os menores limites de crédito, não existem clientes com cartão do tipo platinum e há predominância do gênero feminino;
*   Os clientes com maiores limites de crédito são em sua maioria homens;
*   Os clientes com menores limites de crédito são em sua maioria mulheres;
*   A faixa salarial dos clientes influencia diretamente no limite de crédito obtido;
*   Não existem clientes com salário anual acima de 60k que sejam mulheres.

De acordo com as informações observadas neste dataset, analisamos que **a faixa salarial tem impacto direto no limite de crédito obtido com o banco**, devido a isso, podemos presumir que a**s mulheres tem os menores limites de crédito pois recebem salários menores que os homens**, mesmo atuando nas mesmas funções e com a mesma escolaridade. Uma situação cultural que existe e necessita ser repensada.
